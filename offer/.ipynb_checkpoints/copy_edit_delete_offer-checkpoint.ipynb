```python
import time
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

def login(driver, username, password):
    # Open the website
    driver.get("https://dev.synergi.ai/")
    driver.maximize_window()

    # Enter username
    driver.find_element("id", "email").send_keys(username)

    # Enter password
    driver.find_element("id", "password").send_keys(password)

    # Click on the login button
    driver.find_element(By.CSS_SELECTOR, ".MuiButton-containedSecondary").click()
    WebDriverWait(driver, 20).until(EC.url_changes)
    time.sleep(10)


def navigate_to_offer(driver):
    # Wait for load Navigate to a Offer section
    WebDriverWait(driver, 20).until(EC.presence_of_element_located((By.XPATH, "//span[normalize-space()='Offers']")))
    
    # Navigate to a Offer section
    driver.find_element(By.XPATH, "//span[normalize-space()='Offers']").click()
    time.sleep(2)

def review_save(driver):
    # Wait for load Review
    WebDriverWait(driver, 20).until(EC.element_to_be_clickable((By.CSS_SELECTOR, ".MuiButton-containedSizeMedium.css-15fbcyu")))
    
    # Review
    driver.find_element(By.CSS_SELECTOR, ".MuiButton-containedSizeMedium.css-15fbcyu").click()
    time.sleep(4)
    
    # Wait for load Save Offer
    WebDriverWait(driver, 20).until(EC.element_to_be_clickable((By.CSS_SELECTOR, ".MuiButton-containedSizeMedium.css-1hw9j7s")))
   
    # Save Offer
    driver.find_element(By.CSS_SELECTOR, ".MuiButton-containedSizeMedium.css-1hw9j7s").click()              
    time.sleep(4)


def copy_offer(driver):
    # Select copy
    WebDriverWait(driver, 20).until(EC.element_to_be_clickable((By.CSS_SELECTOR, "div:nth-child(2) > button:nth-child(2)")))
    driver.find_element(By.CSS_SELECTOR, "div:nth-child(2) > button:nth-child(2)").click()
        
    review_save(driver)

def edit_offer(driver):
    # Select Edit
    WebDriverWait(driver, 20).until(EC.element_to_be_clickable((By.CSS_SELECTOR, "button:nth-child(5)")))
    driver.find_element(By.CSS_SELECTOR, "button:nth-child(5)").click()
        
    review_save(driver)
    
def delete_offer(driver):
    # Select Delete img
    WebDriverWait(driver, 20).until(EC.element_to_be_clickable((By.CSS_SELECTOR, "button:nth-child(4) > img")))
    driver.find_element(By.CSS_SELECTOR, "button:nth-child(4) > img").click()

    # Select Delete 
    WebDriverWait(driver, 20).until(EC.element_to_be_clickable((By.CSS_SELECTOR, "div:nth-child(3) > button:nth-child(2)")))
    driver.find_element(By.CSS_SELECTOR, "div:nth-child(3) > button:nth-child(2)").click()
    time.sleep(2)

# ced for copy edit delete offer
def cedo():
    try:
        # Create a webdriver instance
        driver = webdriver.Chrome()

        # login credentials
        username = "ketulpadariya79@gmail.com"
        password = "Ketul@123"

        login(driver, username, password)
        time.sleep(2)
        navigate_to_offer(driver)
        time.sleep(2)
        copy_offer(driver)
        time.sleep(4)
        edit_offer(driver)
        time.sleep(4)
        delete_offer(driver)
        time.sleep(4)

    except Exception as e:
        print(e)
        

# Call the function
cedo()
```
