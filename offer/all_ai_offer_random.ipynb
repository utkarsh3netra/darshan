```python
import time
import random
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

def _login(driver, username, password):
    # Enter username
    driver.find_element("id", "email").send_keys(username)

    # Enter password
    driver.find_element("id", "password").send_keys(password)

    # Click on the login button
    driver.find_element(By.CSS_SELECTOR, ".MuiButton-containedSecondary").click()

    # Wait for successful login
    WebDriverWait(driver, max).until(EC.url_changes)
    time.sleep(8)

max = 60

def navigate_to_offer(driver):
    # Navigate to a offer section
    WebDriverWait(driver, max).until(EC.presence_of_element_located((By.CSS_SELECTOR, "a:nth-child(4) > div:nth-child(1)")))
    driver.find_element(By.CSS_SELECTOR, "a:nth-child(4) > div:nth-child(1)").click()
    time.sleep(2)


def create_offer(driver):
    # Create offer
    time.sleep(2)
    WebDriverWait(driver, max).until(EC.presence_of_element_located((By.CSS_SELECTOR, "a:nth-child(2) > div:nth-child(2) > span:nth-child(1)")))
    driver.find_element(By.CSS_SELECTOR, "a:nth-child(2) > div:nth-child(2) > span:nth-child(1)").click()
    time.sleep(2)
    
def offer_name():
    for i in range(4):
        Offer_name=random.randint(1,9999)
    return Offer_name
    
def edit_offer(driver, Offer_name):
    # Select Edit
    WebDriverWait(driver, max).until(EC.presence_of_element_located((By.CSS_SELECTOR, ".MuiButton-outlinedSizeMedium.css-cz5i37")))
    driver.find_element(By.CSS_SELECTOR, ".MuiButton-outlinedSizeMedium.css-cz5i37").click()

    # Enter offer name
    driver.find_element("id", "offerName").send_keys(Offer_name)

    # Review
    WebDriverWait(driver, max).until(EC.element_to_be_clickable((By.CSS_SELECTOR, ".MuiButton-containedSizeMedium.css-15fbcyu")))
    driver.find_element(By.CSS_SELECTOR, ".MuiButton-containedSizeMedium.css-15fbcyu").click()
    
    # Save offer
    WebDriverWait(driver, max).until(EC.element_to_be_clickable((By.CSS_SELECTOR, ".MuiButton-containedSizeMedium.css-1hw9j7s")))
    driver.find_element(By.CSS_SELECTOR, ".MuiButton-containedSizeMedium.css-1hw9j7s").click()
    time.sleep(4)

def general(driver):
    create_offer(driver)

    # Enter offer name for editing
    Offer_name=offer_name()
        
    # Create General
    WebDriverWait(driver, max).until(EC.presence_of_element_located((By.CSS_SELECTOR, "div:nth-child(4) button:nth-child(1)")))
    driver.find_element(By.CSS_SELECTOR, "div:nth-child(4) button:nth-child(1)").click()

    edit_offer(driver, Offer_name)


def fha_pi_savings(driver):
    create_offer(driver)

    # Enter offer name for editing
    Offer_name=offer_name()
        
    # Create FHA PI Savings
    WebDriverWait(driver, max).until(EC.presence_of_element_located((By.CSS_SELECTOR, ":nth-child(2) >.MuiCardActions-root.MuiCardActions-spacing.css-3zukih button")))
    driver.find_element(By.CSS_SELECTOR, ":nth-child(2) >.MuiCardActions-root.MuiCardActions-spacing.css-3zukih button").click()

    edit_offer(driver, Offer_name)


def rate_term(driver):
    create_offer(driver)

    # Enter offer name for editing
    Offer_name=offer_name()

    # wait for load Create Rate Term PI Benefit
    WebDriverWait(driver, max).until(EC.presence_of_element_located((By.CSS_SELECTOR, " :nth-child(3) > :nth-child(4) :nth-child(1)")))
        
    # Create Rate Term PI Benefit
    driver.find_element(By.CSS_SELECTOR, " :nth-child(3) > :nth-child(4) :nth-child(1)").click()

    edit_offer(driver, Offer_name)


def cashout(driver):
    create_offer(driver)

    # Enter offer name for editing
    Offer_name=offer_name()

    #slid offer
    driver.find_element(By.XPATH, "//button[2]").click()
    time.sleep(2)

    # wait for load Create first_time_freddy
    WebDriverWait(driver, max).until(EC.presence_of_element_located((By.XPATH, "//div[4]//div[4]//button[1]")))

    # Create first_time_freddy
    driver.find_element(By.XPATH, "//div[4]//div[4]//button[1]").click()

    edit_offer(driver, Offer_name)


def purchase_fha(driver):
    create_offer(driver)

    # Enter offer name for editing
    Offer_name=offer_name()

    #slid offer
    driver.find_element(By.XPATH, "//button[2]").click()
    time.sleep(2)

    # wait for load Create Purchase FHA
    WebDriverWait(driver, max).until(EC.presence_of_element_located((By.CSS_SELECTOR, "div:nth-child(5) div:nth-child(4) button:nth-child(1)")))
        
    # Create Purchase FHA
    driver.find_element(By.CSS_SELECTOR, "div:nth-child(5) div:nth-child(4) button:nth-child(1)").click()

    edit_offer(driver, Offer_name)


def login_offer(env ,*,username="ketulpadariya79@gmail.com",password="Krupali@123"):
    driver = webdriver.Chrome()
    driver.maximize_window()

    
    site_subdomain = "app" if env.lower().startswith("prod") else "dev"

    if env == "dev":
        username == "ketulpadariya79@gmail.com"
        password = "Ketul@123"
    else :
        pass
        
    try:  
        # Open the website
        driver.get(f"https://{site_subdomain}.synergi.ai/")


        # Call login function
        _login(driver, username, password)
            
        navigate_to_offer(driver)
        time.sleep(2)
        general(driver)
        time.sleep(2)
        fha_pi_savings(driver)
        time.sleep(2)
        rate_term(driver)
        time.sleep(2)
        cashout(driver)
        time.sleep(2)
        purchase_fha(driver)
        time.sleep(2)

    except Exception as e:
        print(e)
        return driver

```
