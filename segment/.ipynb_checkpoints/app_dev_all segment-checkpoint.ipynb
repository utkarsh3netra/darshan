```python
import time
import random
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

def _login(driver,username, password):
    # login credentials
    #username = "ketulpadariya79@gmail.com"
    #password = "Ketul@123"
    
    # Enter username
    driver.find_element("id", "email").send_keys(username)

    # Enter password
    driver.find_element("id", "password").send_keys(password)

    # Click on the login button
    driver.find_element(By.CSS_SELECTOR, ".MuiButton-containedSecondary").click()
    WebDriverWait(driver, 20).until(EC.url_changes)
    time.sleep(10)

    return driver


def navigate_to_segments(driver):
    # Wait for load Navigate to a Segment section
    WebDriverWait(driver, 20).until(EC.presence_of_element_located((By.XPATH, "//span[normalize-space()='Segments']")))
    
    # Navigate to a Segment section
    driver.find_element(By.XPATH, "//span[normalize-space()='Segments']").click()
    time.sleep(2)


def create_segment(driver):
    # Wait for load Create Segment
    WebDriverWait(driver, 20).until(EC.presence_of_element_located((By.XPATH, "//button[normalize-space()='create Segment']")))
    
    # Create Segment
    driver.find_element(By.XPATH, "//button[normalize-space()='create Segment']").click()
    time.sleep(2)
    
def segment_name():
    for i in range(4):
        Segment_name=random.randint(1,9999)
    return Segment_name

def edit_segment(driver, Segment_name):
    # Wait for load Select Edit
    WebDriverWait(driver, 20).until(EC.presence_of_element_located((By.XPATH, "//button[normalize-space()='Edit']")))
    
    # Select Edit
    driver.find_element(By.XPATH, "//button[normalize-space()='Edit']").click()

    # Enter segment name
    driver.find_element("id", "segmentName").send_keys(Segment_name)
    time.sleep(2)

    # Wait for load Review
    WebDriverWait(driver, 20).until(EC.element_to_be_clickable((By.XPATH, "//button[normalize-space()='Review']")))

    # Review
    driver.find_element(By.XPATH, "//button[normalize-space()='Review']").click()
    time.sleep(10)
    
    # Wait for load Save Segment
    WebDriverWait(driver, 20).until(EC.element_to_be_clickable((By.XPATH, "//button[normalize-space()='Save Segment']")))
   
    # Save Segment
    driver.find_element(By.XPATH, "//button[normalize-space()='Save Segment']").click()              
    time.sleep(10)

    # Wait for load Save Manage Segment
    WebDriverWait(driver, 20).until(EC.element_to_be_clickable((By.CSS_SELECTOR, "button[class='MuiButton-root']")))
    
    # Manage Segment
    driver.find_element(By.CSS_SELECTOR, "button[class='MuiButton-root']").click()
    time.sleep(2)


def refinancer_heloc(driver):
    create_segment(driver)

    # segment name
    Segment_name = segment_name()

    # wait for load Create Conservative Carla
    WebDriverWait(driver, 20).until(EC.presence_of_element_located((By.CSS_SELECTOR, "div:nth-child(4) button:nth-child(1)")))

    # Create Conservative Carla
    driver.find_element(By.CSS_SELECTOR, "div:nth-child(4) button:nth-child(1)").click()
        
    edit_segment(driver, Segment_name)


def refinancer_refinance(driver):
    create_segment(driver)

    # segment name
    Segment_name = segment_name()

    # wait for load Create Investor Ian
    WebDriverWait(driver, 20).until(EC.presence_of_element_located((By.XPATH, "//div[2]/div[4]/button")))

    # Create Investor Ian
    driver.find_element(By.XPATH, "//div[2]/div[4]/button").click()
    
    edit_segment(driver, Segment_name)


def downsizer(driver):
    create_segment(driver)

    # segment name 
    Segment_name = segment_name()

    # wait for load Create Downsizer 
    WebDriverWait(driver, 20).until(EC.presence_of_element_located((By.XPATH, "//div[3]//div[4]//button[1]")))

    # Create Downsizer Dave
    driver.find_element(By.XPATH, "//div[3]//div[4]//button[1]").click()

    edit_segment(driver, Segment_name)


def investor_cashout(driver):
    create_segment(driver)

    # segment name 
    Segment_name = segment_name()

    #slid arrow
    driver.find_element(By.XPATH, "//button[2]").click()
    time.sleep(2)

    # wait for load Create investor_cashout
    WebDriverWait(driver, 20).until(EC.presence_of_element_located((By.XPATH, "//div[4]//div[4]//button[1]")))

    # Create investor_cashout
    driver.find_element(By.XPATH, "//div[4]//div[4]//button[1]").click()

    edit_segment(driver, Segment_name)


def conservative_refinance(driver):
    create_segment(driver)

    # segment name
    Segment_name = segment_name()

    #slid arrow
    driver.find_element(By.XPATH, "//button[2]").click()
    time.sleep(2)
    driver.find_element(By.XPATH, "//button[2]").click()
    time.sleep(2)

    # wait for load Create conservative_refinance
    WebDriverWait(driver, 20).until(EC.presence_of_element_located((By.XPATH, "//div[5]//div[4]//button[1]")))

    # Create conservative_refinance
    driver.find_element(By.XPATH, "//div[5]//div[4]//button[1]").click()

    edit_segment(driver, Segment_name)

def conservative_heloc(driver):
    create_segment(driver)

    # segment name
    Segment_name = segment_name()

    #slid arrow
    driver.find_element(By.XPATH, "//button[2]").click()
    time.sleep(2)
    driver.find_element(By.XPATH, "//button[2]").click()
    time.sleep(2)

    # wait for load Create conservative_heloc
    WebDriverWait(driver, 20).until(EC.presence_of_element_located((By.XPATH, "//div[6]//div[4]//button[1]")))

    # Create conservative_heloc
    driver.find_element(By.XPATH, "//div[6]//div[4]//button[1]").click()

    edit_segment(driver, Segment_name)

def refinancer_cashout(driver):
    create_segment(driver)

    # segment name
    Segment_name = segment_name()

    #slid arrow
    driver.find_element(By.XPATH, "//button[2]").click()
    time.sleep(2)
    driver.find_element(By.XPATH, "//button[2]").click()
    time.sleep(2)
    driver.find_element(By.XPATH, "//button[2]").click()
    time.sleep(2)

    # wait for load Create refinancer_cashout
    WebDriverWait(driver, 20).until(EC.presence_of_element_located((By.XPATH, "//div[7]//div[4]//button[1]")))

    # Create refinancer_cashout
    driver.find_element(By.XPATH, "//div[7]//div[4]//button[1]").click()

    edit_segment(driver, Segment_name)

def upgrader_purchase_fha(driver):
    create_segment(driver)

    # segment name
    Segment_name = segment_name()

    #slid arrow
    driver.find_element(By.XPATH, "//button[2]").click()
    time.sleep(2)
    driver.find_element(By.XPATH, "//button[2]").click()
    time.sleep(2)
    driver.find_element(By.XPATH, "//button[2]").click()
    time.sleep(2)
    driver.find_element(By.XPATH, "//button[2]").click()
    time.sleep(2)
    
    # wait for load Create upgrader_purchase_fha
    WebDriverWait(driver, 20).until(EC.presence_of_element_located((By.XPATH, "//div[8]//div[4]//button[1]")))

    # Create upgrader_purchase_fha
    driver.find_element(By.XPATH, "//div[8]//div[4]//button[1]").click()

    edit_segment(driver, Segment_name)

def upgrader_cashout(driver):
    create_segment(driver)

    # segment name
    Segment_name = segment_name()

    #slid arrow
    driver.find_element(By.XPATH, "//button[2]").click()
    time.sleep(2)
    driver.find_element(By.XPATH, "//button[2]").click()
    time.sleep(2)
    driver.find_element(By.XPATH, "//button[2]").click()
    time.sleep(2)
    driver.find_element(By.XPATH, "//button[2]").click()
    time.sleep(2)
    driver.find_element(By.XPATH, "//button[2]").click()
    time.sleep(2)

    # wait for load Create upgrader_cashout
    WebDriverWait(driver, 20).until(EC.presence_of_element_located((By.XPATH, "//div[9]//div[4]//button[1]")))

    # Create upgrader_cashout
    driver.find_element(By.XPATH, "//div[9]//div[4]//button[1]").click()

    edit_segment(driver, Segment_name)

def segment():
    try:
        driver = webdriver.Chrome()
        
        navigate_to_segments(driver)
        time.sleep(2)
        
        refinancer_heloc(driver)
        time.sleep(2)
        refinancer_refinance(driver)
        time.sleep(2)
        downsizer(driver)
        time.sleep(2)
        investor_cashout(driver)
        time.sleep(2)
        conservative_refinance(driver)
        time.sleep(2)
        conservative_heloc(driver)
        time.sleep(2)
        refinancer_cashout(driver)
        time.sleep(2)
        upgrader_purchase_fha(driver)
        time.sleep(2)
        upgrader_cashout(driver)
        time.sleep(2)
    
    except Exception as e:
        print(e)
        
def login_to_the_site(env ,*,username="ketulpadariya79@gmail.com",password="Krupali@123"):
    driver = webdriver.Chrome()
    driver.maximize_window()

    
    site_subdomain = "app" if env.lower().startswith("prod") else "dev"

    if env == "dev":
        username == "ketulpadariya79@gmail.com"
        password = "Ketul@123"
    else :
        pass
        
    try:  
        # Open the website
        driver.get(f"https://{site_subdomain}.synergi.ai/")

    
    
        #Call login function
        _login(driver, username, password)

    
        
        navigate_to_segments(driver)
        time.sleep(2)
        
        refinancer_heloc(driver)
        time.sleep(2)
        refinancer_refinance(driver)
        time.sleep(2)
        downsizer(driver)
        time.sleep(2)
        investor_cashout(driver)
        time.sleep(2)
        conservative_refinance(driver)
        time.sleep(2)
        conservative_heloc(driver)
        time.sleep(2)
        refinancer_cashout(driver)
        time.sleep(2)
        upgrader_purchase_fha(driver)
        time.sleep(2)
        upgrader_cashout(driver)
        time.sleep(2)
    
    except Exception as e:
         return driver
         print(e)
```
